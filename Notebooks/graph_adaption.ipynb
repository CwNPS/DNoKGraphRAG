- GraphReasoning currently assumes the following
    - undirected graph
    - node_ids are the label/keyword for that node
    - edges have "title" field that describes the relationship between nodes

```python
#load graph file
graph_path = ".\\data\\generated_graphs\\SME_graph\\DNoKv3.graphml"
graph = nx.read_graphml(graph_path)
```

```python
def update_node_id(G, old_id):
    """ 
    This function updates node ids to the title information so that the current embedding and node search functions work. 
    """
    new_id = G.nodes[old_id]['label']

    #print(f"Updating node ID from '{old_id}' to '{new_id}'")
    # Check if the old node exists
    if old_id not in G :
        return False
    
    if new_id in G:
        #options merge nodes or append label so that new id is not the same

        # ## merge option: 
        # def merge_nodes(G, old_id, new_id, attributes):
        #     for key, value in attributes.items():
        #         if key in G.nodes[old_id]:
        #             G.nodes[old_id][key] = f"{G.nodes[old_id][key]}, {value}"
        #         else:
        #             G.nodes[old_id][key] = value
        #     # Reconnect edges
        #     for neighbor in G.neighbors(new_id):
        #         if not G.has_edge(old_id, neighbor):
        #             G.add_edge(old_id, neighbor, **G.edges[new_id, neighbor])
        #     G.remove_node(new_id)

        ##append option
        new_id= f"{new_id} from {G.nodes[old_id]['title']}"


    # Create the new node with attributes from the old node
    G.add_node(new_id, **G.nodes[old_id])

    if graph.is_directed():
        # Reconnect all edges to the new node
        for predecessor in G.predecessors(old_id):
            G.add_edge(predecessor, new_id, **G.edges[predecessor, old_id])
        for successor in G.successors(old_id):
            G.add_edge(new_id, successor, **G.edges[old_id, successor])
    else:
        # Reconnect all edges to the new node
        for neighbor in G.neighbors(old_id):
            G.add_edge(neighbor, new_id, **G.edges[neighbor, old_id])

    # Remove the old node
    G.remove_node(old_id)
    # Check if the old node was successfully removed
    if old_id in G:
        print(f"Node update failed: old_id '{old_id}' was not removed.")
        return False

    return True
```

```python
#change node_ids to to label for embedding
for node in list(graph.nodes()):
    update_node_id(graph, node)

print_graph_info(graph)


# Add the 'title' attribute to all edges. Can not determine specific relationships yet.
nx.set_edge_attributes(graph, 'has keywords in common with', 'title')
print_graph_info(graph)
```
